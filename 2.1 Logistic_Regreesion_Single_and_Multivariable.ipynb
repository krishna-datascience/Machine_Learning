# Logistic_Regression_single_variable

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("insurance_data.csv")
df
```

```python
plt.scatter(df.age,df.bought_insurance)
```

```python
df.hist(bins = 40)
```

```python
from sklearn.linear_model import LogisticRegression
#Logistic_regression_used_to_make_linear_classification

reg= LogisticRegression()
reg.fit(df[["age"]],df.bought_insurance)
```

```python
reg.predict([[34],[24],[56],[30],[31],[32],[33],[28],[78]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(df[["age"]],df.bought_insurance)
```

#MUTIVARIABLE_LOGISTIC_REGRESSION

```python
df1 = pd.read_csv("HR.csv")
df1.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
label.fit_transform(df1.Department)
```

```python
df1.Department = label.fit_transform(df1.Department)
```

```python
label.fit_transform(df1.salary)
```

```python
df1.salary = label.fit_transform(df1.salary)
```

```python
df1.head(3)
```

```python
df1.describe()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
import seaborn as sn
sn.pairplot(df1, hue ="salary")
```

```python
sn.pairplot(df1, hue ="salary",kind = "reg")
```

```python
df1.info()
```

```python
x = df1.drop("salary",axis = 1)
y = df1.salary
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =   train_test_split(x,y,test_size = 0.3,random_state = 0)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
reg1 = LogisticRegression()
reg1.fit(x_train,y_train)
```

```python
reg1.predict(x_test)
```

```python
reg1.score(x_train,y_train)
```

```python
reg1.score(x_test,y_test)
```

```python
y_predicted = reg1.predict(x_test)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
plt.figure(figsize=(8,5))
sn.heatmap(cm,annot = True)
plt.xlabel("PREDICTED")
plt.ylabel("TRUTH_OR_ACTUAL_VALUE")
```
