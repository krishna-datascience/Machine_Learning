# UNSUPERVISED LEARNING : -


* What is Unsupervised Learning?

 Unsupervised learning is a machine learning technique, where you do not need to supervise the model. Instead, you need to allow the model to work on its own to discover information. It mainly deals with the unlabelled data.
 
 Unsupervised learning algorithms allows you to perform more complex processing tasks compared to supervised learning. Although, unsupervised learning can be more unpredictable compared with other natural learning methods.

```python
import pandas as pd

import matplotlib.pyplot as plt

data = pd.read_csv("Customers.csv")
data.head()
```

```python
plt.scatter(data["Annual Income (k$)"],data["Spending Score (1-100)"])
```

```python
x = data.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans

km = KMeans()
```

```python
wcss = [] # with in cluster sum of square wcss = sum(c_i + x_i)**2
 
for i  in range(1,15):
    km = KMeans(i)
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,15),wcss)
plt.xlabel("K")
plt.ylabel("SUM_OF_SQUARE")
```

```python
x
```

```python
km = KMeans(5,random_state=0) # which is used to initialize the centeroids of cluster 
y_km = km.fit_predict(x)
y_km
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x[y_km == 0,0],x[y_km == 0,1],label = "c1")
plt.scatter(x[y_km == 1,0],x[y_km == 1,1],label = "c2")
plt.scatter(x[y_km == 2,0],x[y_km == 2,1],label = "c3")
plt.scatter(x[y_km == 3,0],x[y_km == 3,1],label = "c4")
plt.scatter(x[y_km == 4,0],x[y_km == 4,1],label = "c5")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker = "*",s =234,color = "black",label = "Centeroids")
plt.legend()
```

```python
df  = pd.read_csv("loan_data.csv")
df.head()
```

```python
d = df[["revol.util","fico"]]
```

```python
d.head()
```

```python
sse = []
for i in range(1,15):
    km1 = KMeans(n_clusters=i)
    km1.fit(d)
    sse.append(km1.inertia_)
```

```python
sse
```

```python
plt.plot(range(1,15),sse)
plt.xlabel("K")
plt.ylabel("SSE")
```

```python
km1 = KMeans(5)
km1.fit_predict(d)
```

```python
d["cluster"] = km1.fit_predict(d)
```

```python
d.head()
```

```python
c1 = d[d.cluster ==0]
c2 = d[d.cluster ==1]
c3 = d[d.cluster ==2]
c4 = d[d.cluster ==3]
c5 = d[d.cluster ==4]
```

```python
c4
```

```python
plt.figure(figsize=(10,8))
plt.scatter(c1["revol.util"],c1["fico"],label = "c1")
plt.scatter(c2["revol.util"],c2["fico"],label = "c2")
plt.scatter(c3["revol.util"],c3["fico"],label = "c3")
plt.scatter(c4["revol.util"],c4["fico"],label = "c4")
plt.scatter(c5["revol.util"],c5["fico"],label = "c5")
plt.scatter(km1.cluster_centers_[:,0],km1.cluster_centers_[:,1],label = "centroids",marker = "*",color = "black",s = 234)
plt.legend()
```
