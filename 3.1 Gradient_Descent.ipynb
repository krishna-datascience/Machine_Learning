```python
import numpy as np

def gradient_descent(x,y):
    m=b=0
    learning_rate = 0.001
    iteration = 1000
    n = len(x)
    for i in range(iteration):
        
        y_prediction = m * x + b
        cf = (1/n)*sum(y-y_prediction)**2
        md = (-2/n)*sum(x*(y - y_prediction))
        bd = (-2/n)*sum(y-y_prediction)
        m = m - learning_rate*md
        b  = b- learning_rate*bd
        print("iteration {} : ,coef{}:,inter_ {}: ,cost{}".format(i,m,b,cf))
        
x = np.array([1,2,3,4,5])
y = np.array([1,3,5,7,11])
gradient_descent(x,y)
```
